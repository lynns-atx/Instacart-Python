## 2. Follow Exercise instructions to create a bar chart, a histogram and a line chart

### Bar Chart

```python
# Create a bar chart

all['orders_day_of_week'].value_counts().plot.bar()
```

### Line Chart

```python
# Create a sample of the population

# Create a list holding True/False values to the test np.random.rant() <+ 0.7

np.random.seed(4)
dev = np.random.rand(len(all)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = all[dev]
```

```python
# Store 30% of the sample in the data frame small

small = all[~dev]
```

```python
# Check numbers to see if they are correct

len(all)
```

```python
len(big) + len(small)
```

```python
# Only including the relevant columns

df_2 = small[['orders_day_of_week','prices']]
```

```python
# Create line chart

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export line chart as an image

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_prices_day.png'))
```
